# Credit Card Fraud Detection Visualization
This notebook contains visualizations for analyzing the credit card fraud detection model's performance, including:
- Class distribution comparisons between actual and predicted fraud cases
- Feature distributions for key transaction attributes
- Probability distributions showing model confidence
- ROC curves and other performance metrics

Make sure you have all required packages installed from requirements.txt before running this notebook

### Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import sys
sys.path.append('../src')
from utils.helpers import plot_class_distribution, plot_feature_distribution
import seaborn as sns
```

### Function Description
#### The visualize_results() function below performs the following tasks:
1. Loads the original credit card transaction data and model predictions
2. Creates visualizations for:
    - Original class distribution (fraud vs non-fraud)
    - Predicted class distribution from the model
    - Distribution plots for important transaction features (V1-V4, Amount)
    - Histogram of fraud probability scores
#### You also can save all plots to the predictions directory for later analysis

```python
def visualize_results():
    # Load the original data
    print("Loading data...")
    df = pd.read_csv("../data/creditcard.csv")
    
    # Load predictions
    print("Loading predictions...")
    predictions = pd.read_csv("../predictions/predictions.csv")
    
    # 1. Original Class Distribution
    print("\nPlotting original class distribution...")
    plt.figure(figsize=(10, 6))
    plot_class_distribution(df['Class'].values)
    plt.close()
    
    # 2. Predicted Class Distribution
    print("Plotting predicted class distribution...")
    plt.figure(figsize=(10, 6))
    plot_class_distribution(predictions['Predicted_Class'].values)
    plt.close()
    
    # 3. Feature Distributions for key features
    print("\nPlotting feature distributions...")
    important_features = ['V1', 'V2', 'V3', 'V4', 'Amount']  # You can modify this list
    for feature in important_features:
        if feature in df.columns:
            print(f"Plotting distribution for {feature}...")
            plot_feature_distribution(df, feature)
            plt.close()
    
    # 4. Probability Distribution for Fraud Cases
    print("\nPlotting probability distributions...")
    plt.figure(figsize=(10, 6))
    sns.histplot(data=predictions, x='Probability_Class_1', bins=50)
    plt.title('Distribution of Fraud Probabilities')
    plt.xlabel('Probability of Fraud')
    plt.ylabel('Count')
    plt.savefig('../predictions/fraud_probability_distribution.png')
    plt.close()
    
    print("\nVisualization complete! Check the 'predictions' directory for the generated plots.")
```

```python
if __name__ == "__main__":
    visualize_results() 
```
